# Dane steps-large

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("../Data/regression/steps-large-training.csv", index_col = 0).to_numpy()
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
```

```python
# === Define MLP Model ===
class MLP(nn.Module):
    def __init__(self, input_size, hidden_layers, output_size, activation_fn=nn.Sigmoid, output_activation=nn.Identity):
        """
        input_size: Number of input features
        hidden_layers: List of integers, each defining the number of neurons in a hidden layer
        output_size: Number of output neurons
        activation_fn: Activation function for hidden layers (default: Sigmoid)
        output_activation: Activation function for output layer (default: Linear/Identity)
        """
        super(MLP, self).__init__()
        
        layers = []
        prev_size = input_size
        
        for hidden_size in hidden_layers:
            layers.append(nn.Linear(prev_size, hidden_size))
            layers.append(activation_fn())
            prev_size = hidden_size
        
        # Output layer (Linear)
        layers.append(nn.Linear(prev_size, output_size))
        layers.append(output_activation())  # Default is Identity (linear)
        
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        return self.model(x)

# === Load & Preprocess Data ===
def standardize_data(x, mean=None, std=None):
    """ Standardizes data (zero mean, unit variance) """
    if mean is None: mean = np.mean(x, axis=0)
    if std is None: std = np.std(x, axis=0)
    return (x - mean) / std, mean, std

def destandardize_data(y, mean, std):
    """ Reverts standardization """
    return y * std + mean

# Example synthetic dataset
x_train = x.copy()
y_train = y.copy()

# Standardize input/output
x_train_std, x_mean, x_std = standardize_data(x)
y_train_std, y_mean, y_std = standardize_data(y)

# Convert to PyTorch tensors
x_train_tensor = torch.tensor(x_train_std, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_std, dtype=torch.float32)

# === Training Function ===
def train_mlp(model, x_train, y_train, epochs=20000, lr=0.02):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    for epoch in range(epochs):
        optimizer.zero_grad()
        y_pred = model(x_train)
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()

        if epoch % 500 == 0:
            print(f"Epoch {epoch}: MSE = {loss.item():.4f}")

# === Define & Train Models for Different Architectures ===
architectures = [
    [5, 5],     # One hidden layer, 5 neurons
]

for arch in architectures:
    print(f"\nTraining MLP with architecture: {arch}")
    
    model = MLP(input_size=1, hidden_layers=arch, output_size=1)
    train_mlp(model, x_train_tensor, y_train_tensor)

    # Predict & plot results
    with torch.no_grad():
        y_pred_std = model(x_train_tensor).numpy()
        y_pred = destandardize_data(y_pred_std, y_mean, y_std)

    plt.scatter(x_train, y_train, label="True Function", linestyle="dashed", color="black")
    plt.scatter(x_train, y_pred, label=f"MLP {arch}", alpha=0.8)
    plt.legend()
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title(f"MLP Approximation ({arch})")
    plt.show()
```
